```r
# Scatterplot: Age vs Total Played Hours
options(repr.plot.height=6, repr.plot.width= 8)

players_age <- ggplot(players, aes(x = age, y = played_hours)) +
               geom_point() +
               geom_smooth(method = "lm", se = FALSE) +
               labs(title = "Relationship Between Age and Total Played Hours",
                    x = "Age",
                    y = "Total Played Hours") +
            theme(text=element_text(size=15))
players_age
```

```r
set.seed(123)  # For reproducibility

players_split <- initial_split(players, prop= 0.75, strata= played_hours) 
players_training <- training(players_split) 

players_testing <- testing(players_split)

players_recipe <- recipe(played_hours~ experience + age + gender, data = players_training) |>
                  step_dummy(all_nominal(), -all_outcomes()) |>
                  step_scale(all_predictors()) |>
                 step_center(all_predictors()) 

players_spec <- nearest_neighbor(weight_func= "rectangular", neighbors= tune()) |>
                set_engine("kknn") |>
                set_mode("regression")

players_v_fold <- vfold_cv(players_training, v=5, strata= played_hours)

players_workflow <- workflow() |>
                    add_recipe(players_recipe) |>
                    add_model(players_spec)

                    
players_workflow
```

```r
 # Convert logical to numeric

players_grid_results <- tune_grid(players_workflow,
                resamples = players_v_fold,
                grid = 10) 

players_grid_results |>
collect_metrics()
```

```r
players_best_model <- grid_results |>
  select_best(metric = "rmse")

players_best_model
```

```r
final_workflow <- players_workflow |>
finalize_workflow(players_best_model)

final_workflow
```

```r
# Fit the finalized workflow to the training data and test it
final_fit <- final_workflow |>
  last_fit(split = players_split)

# Collect and view metrics for the testing set
final_fit |>
  collect_metrics()

# Collect predictions for further analysis
predictions <- final_fit |>
  collect_predictions()

# View predictions
head(predictions)
```

```r
library(ggplot2)

# Scatterplot of Predicted vs Actual Played Hours
predictions |>
  ggplot(aes(x = .pred, y = played_hours)) +
  geom_point() +
  geom_abline(color = "red", linetype = "dashed") +  # Add a y = x line for reference
  labs(
    title = "Predicted vs Actual Played Hours",
    x = "Predicted Played Hours",
    y = "Actual Played Hours") +
   scale_x_continuous(limits = c(0, 20), breaks = seq(0, 20, 5)) +  
  scale_y_continuous(limits = c(0, 160), breaks = seq(0, 160, 20)) +  
 theme_minimal()

  

```

Interpretation of the Scatterplot
General Pattern:
The points are clustered near the bottom left of the plot, 
indicating that both the predicted and actual played hours are generally low for most players.
The red dashed line (y = x) represents the ideal case where predictions match actual values.
Most points deviate from this line, indicating prediction errors.

Outliers:
There is at least one significant outlier with an actual played hour value above 150. 
The model struggled to predict this value accurately, as the prediction is much lower than the actual value.
Other points with low actual played hours (<50) show better alignment with the predictions.

Model Performance:
The model appears to have a weak predictive performance overall, 
as the predicted values generally underestimate the actual values.
The spread of the points and their distance from the y = x line suggests that 
the model is not accurately capturing the variability in the played_hours response variable.
